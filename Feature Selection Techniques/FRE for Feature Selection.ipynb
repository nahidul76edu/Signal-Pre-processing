```python
# Feature Selection with Univariate Statistical Tests
from pandas import read_csv
import pandas as pd
import numpy as np
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
# Feature Extraction with RFE
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
DataLL10 = pd.read_csv('/content/drive/MyDrive/AEP_ONLINE_DATA/CSV_LL/DataLL10.csv', header= None)
DataRR10 = pd.read_csv('/content/drive/MyDrive/AEP_ONLINE_DATA/CSV_RR/DataRR10.csv', header= None)
```

```python
label1 = []
for i in range(len(DataLL10)):
  label1.append(1)

label2 = []
for i in range(len(DataRR10)):
  label2.append(2)
```

```python
label1 = pd.DataFrame(label1)
DataLL10_L = np.hstack((DataLL10, label1))
DataLL10_W_L = pd.DataFrame(DataLL10_L)

label2 = pd.DataFrame(label2)
DataRR10_L = np.hstack((DataRR10, label2))
DataRR10_W_R = pd.DataFrame(DataRR10_L)
```

```python
DATA_S10 = np.concatenate((DataLL10_W_L, DataRR10_W_R))
```

```python
DATA_S10_f = pd.DataFrame(DATA_S10)
```

```python
#array = Dataset.values
X = DATA_S10_f.iloc[:,0:63]
Y = DATA_S10_f.iloc[:, 64]
```

```python
# feature extraction
model = LogisticRegression(solver='lbfgs')
rfe = RFE(model, 20)
fit = rfe.fit(X, Y)

print("Num Features: %d" % fit.n_features_)
print("Selected Features: %s" % fit.support_)
print("Feature Ranking: %s" % fit.ranking_)
```

```python
data_l10 = []
j = 0
for i in fit.ranking_:
  if i == 1:
    data1 = DataLL10.iloc[:,j]
    data_l10.append(data1)
  j +=1
```

```python
data10L_1 = pd.DataFrame(data_l10)
data10L_2 = data10L_1.T
data10L_L = pd.DataFrame(data10L_2)
```

```python
data_r10 = []
j = 0
for i in fit.ranking_:
  if i == 1:
    data1 = DataRR10.iloc[:,j]
    data_r10.append(data1)
  j +=1
```

```python
data10R_1 = pd.DataFrame(data_r10)
data10R_2 = data10R_1.T
data10R_R = pd.DataFrame(data10R_2)
```
