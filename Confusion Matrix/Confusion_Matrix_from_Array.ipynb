```python
import numpy as np
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

array = ([[ 69,   0,   0,   0],
 [  4,  77,   0,  0],
 [  0,   0, 103,   4],
 [  0,   0,   1, 89]])

font = {'family' : 'normal',
        'size'   : 14}

plt.rc('font', **font)

label = ['Class1', 'Class2', 'Class3', 'Class4']


df_cm = pd.DataFrame(array, index = [i for i in label],
                  columns = [i for i in label])

plt.figure(figsize = (5,3))
sn.heatmap(df_cm, annot=True, fmt='.0f', cmap='Blues')
plt.ylabel('True Class')
plt.xlabel('Predicted Class')
plt.show()
```
